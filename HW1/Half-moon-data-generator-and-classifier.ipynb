```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
import matplotlib
```

```python
x, y = datasets.make_moons(n_samples=1000, noise=0.10)
```

```python
plt.scatter(x[np.where(y==0), 0], x[np.where(y==0), 1])
plt.scatter(x[np.where(y==1), 0], x[np.where(y==1), 1])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver="lbfgs").fit(x, y)
clf.score(x, y)
y_pred = clf.predict(x)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = x[:, 0].min() - .5, x[:, 0].max() + .5
y_min, y_max = x[:, 1].min() - .5, x[:, 1].max() + .5
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
tab10 =plt.get_cmap("tab10")
print(tab10.colors)
newtab10 = tab10(np.linspace(0, 1, 12))
newtab10[1] = newtab10[2]
from matplotlib.colors import ListedColormap
newtab10 = ListedColormap(newtab10)
plt.pcolormesh(xx, yy, Z, cmap=newtab10, alpha=1)

plt.scatter(x[np.where(y==0), 0], x[np.where(y==0), 1])
plt.scatter(x[np.where(y==1), 0], x[np.where(y==1), 1])

plt.show()
```
