## Half moon data generator and linear classifier

```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
```

```python
x, y = datasets.make_moons(n_samples=1000, noise=0.10)
plt.scatter(x[np.where(y==0), 0], x[np.where(y==0), 1])
plt.scatter(x[np.where(y==1), 0], x[np.where(y==1), 1])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver="lbfgs").fit(x, y)
clf.score(x, y)
y_pred = clf.predict(x)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = x[:, 0].min() - .5, x[:, 0].max() + .5
y_min, y_max = x[:, 1].min() - .5, x[:, 1].max() + .5
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, colors=["tab:blue", "tab:orange"], levels=3, alpha=0.5)

plt.scatter(x[np.where(y==0), 0], x[np.where(y==0), 1])
plt.scatter(x[np.where(y==1), 0], x[np.where(y==1), 1])

plt.show()
```

## Comparing classifiers

```python
from scipy.stats import wilcoxon, friedmanchisquare
hw1scores = np.loadtxt("UA-ECE523-EngrAppMLData/data/hw1-scores.txt", delimiter=",")
```

### Wilcoxon signed-rank test

```python
for i in range(np.size(hw1scores, 1)):
    for j in range(0, np.size(hw1scores, 1)):
        if i == j: continue
        w, p = wilcoxon(hw1scores[:, i], hw1scores[:, j], alternative="less")
        print(w, p)
        if p < 0.05:
            print("There is strong evidence that classifier " + str(i) + " does better than classifier " + str(j) + " at the 0.05 level.")
```

### Friedman test

```python
fcs, p = friedmanchisquare(*hw1scores)
print(fcs, p)
if p < 0.05:
    print("There is strong evidence that some classifiers do better than the rest.")
```

## Sampling from a distribution

```python
from numpy.random import rand
```

```python
def f(M, p):
    cdf = np.cumsum(p)
    x = rand(M)
    y = [np.where(x[m] < cdf)[0][0] + 1 for m in range(M)]
    return y
```

```python
p = [0.2, 0.2, 0.2, 0.1, 0.3]
y = f(100000, p)
plt.hist(y, density=True, bins=[1, 2, 3, 4, 5, 6])
plt.ylabel("Probability")
plt.show()
```

To generate the samples according to the PMF p, I first compute the corresponding CDF. Then I generate M random numbers uniformly distributed in the range [0, 1). Then I check where these values fall in the CDF and assign the corresponding event.

The plot above shows the value of the probabilities estimated from 100000 random draws.
